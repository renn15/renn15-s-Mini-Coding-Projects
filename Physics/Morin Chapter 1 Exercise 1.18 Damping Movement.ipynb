```python
import math
import matplotlib.pyplot as plt
import numpy as np

def damping(A, x_init, x1_init, power, dt, time):
    x = [x_init]
    x1 = [x1_init]
    x2 = [-A*(x1[0]**power)]
    frame = 0

    while frame < time/dt:
        buffer = x[frame] + x1[frame]*dt
        buffer1 = x1[frame] + x2[frame]*dt
        buffer2 = -A*(x1[frame]**power)
        x.append(buffer)
        x1.append(buffer1)
        x2.append(buffer2)  
        frame += 1
        
    print(x[-1])
    
    plt.plot(np.arange(0, (frame)*dt+dt, dt), x, label="Displacement")
    plt.plot(np.arange(0, (frame)*dt+dt, dt), x1, label="Velocity")
    plt.plot(np.arange(0, (frame)*dt+dt, dt), x2, label="Accelaration")
    plt.legend(loc="lower right")
    plt.grid()
    
```

```python
damping(1, 0, 2, 1, 0.001, 20)
```

```python
damping(1, 0, 4, 1, 0.001, 20)
```

For A = 1 with power of one, displacement approaches the original velocity

```python
damping(1, 0, 2, 2, 0.001, 1000)
```

The displacement plot does look like a log function

```python
damping(1, 0, 2, 3, 0.001, 100)
```

The velocity and accelaration would nearly approach zero, while the displacement also slowly approaches a value, resembling a logarithm graph.
